```python
import cv2
import pytesseract
from PIL import Image
```

```python
Image.open('Sample_quote.jpg')
```

```python
img = cv2.imread('Sample_quote.jpg')
```

```python
img_grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
img_thres = cv2.threshold(img_grey, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]
```

```python
img_noise = cv2.medianBlur(img_thres, 5)
```
