```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('fake_reg.csv')
```

```python
df
```

```python
sns.pairplot(df)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['feature1','feature2']].values
y = df['price']
```

```python
y
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.33, random_state=0)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_train.max()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
help(Sequential)
```

```python
model = Sequential()
```

```python
model.add(Dense(4, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(1))

model.compile(optimizer='rmsprop',loss='mse')
```

```python
model.fit(x=X_train,y=y_train,epochs=300)
```

```python
lossdf = pd.DataFrame(model.history.history)
```

```python
lossdf.plot()
```

```python
scaler = MinMaxScaler()
```

```python
X_test = scaler.fit_transform(X_test)
```

```python
model.evaluate(X_test,y_test,verbose=0)
```

```python
model.evaluate(X_train,y_train,verbose=0)
```

```python
test_prediction = model.predict(X_test)
test_prediction
```

```python
test_prediction = pd.Series(test_prediction.reshape(330,))
```

```python
y_truevalues = pd.Series(y_test.values.reshape(330 ,))
```

```python
pred_df = pd.concat([test_prediction,y_truevalues],axis=1)
```

```python
pred_df.columns = ['True y', 'Model preditions']
```

```python
sns.scatterplot(x='True y',y='Model preditions',data=pred_df)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
pred_df['Model preditions']
```

```python
mean_absolute_error(pred_df['True y'],pred_df['Model preditions'])
```

```python
mean_squared_error(pred_df['True y'],pred_df['Model preditions'])**0.5
```
