```python
str.format?
```

```python
"The sum of 1 + 2 is {0}".format(1+2)
```

```python
# .2f first string can be used to represent no. of floating point we need
print("The sum of 1 + 2 is {0:.3f}".format(1/2)) 
print("The sum of 1 + 2 is {0:.2f}".format(1/2))
```

Alternatively to format

```python
f'The sum of 1 + 2 is {1+2}' # the f is called f-strings (short for formatted string literals)
```

## Notes

String literals that are part of a single expression and have only whitespace between them will be implicitly converted to a single string literal. That is, ("spam " "eggs") == "spam eggs".

```python
print('Arun ' 'Mozhi')
```

```python
('Arun ' 'Mozhi') == 'Arun Mozhi'
```
